### Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
### Calculate the average price AND the profit/loss percentages over the course of last -
### 1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
### {Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def avg_price(N):
    return (df['Average Price'].tail(N).sum())


print('Average Price for last N days are as follows : ')
print('Last 1 week', avg_price(5))
print('Last 2 weeks', avg_price(10))
print('Last 1 month', avg_price(20))
print('Last 3 month', avg_price(60))
print('Last 6 month', avg_price(120))
print('Last 1 year', avg_price(240))
```

```python
def profit_loss(N):
    diff = (df['Close Price'].tail(N).iloc[N-1] -
            df['Close Price'].tail(N).iloc[0])
    if diff < 0:
        loss = -(diff)
        loss_per = (loss/df['Close Price'].tail(N).iloc[N-1])*100
        return loss_per
    if diff > 0:
        profit = diff
        profit_per = (profit/df['Close Price'].tail(N).iloc[N-1])*100
        return profit_per


print('Loss/Profit Percentage for last N days are as follows : ')
print('Last 1 week', profit_loss(5))
print('Last 2 weeks', profit_loss(10))
print('Last 1 month', profit_loss(20))
print('Last 3 month', profit_loss(60))
print('Last 6 month', profit_loss(120))
print('Last 1 year', profit_loss(240))
```

### Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. You will note that the first entry of the column will have a ‘Nan’ value. Why does this happen? Either remove the first row, or set the entry to 0 before proceeding.

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change().fillna(0)
df[['Date', 'Day_Perc_Change']]
```

### Add another column 'Trend' whose values are: 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1 'Positive' for 'Day_Perc_Change' in between 1 and 3 'Negative' for 'Day_Perc_Change' in between -1 and -3 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7 'Among top losers' for 'Day_Perc_Change' in between -3 and -7 'Bull run' for 'Day_Perc_Change' >7 'Bear drop' for 'Day_Perc_Change' <-7

```python
if((df['Day_Perc_Change'] >= -0.5) & (df['Day_Perc_Change'] <= 0.5)).all():
    df['Trend'] = 'Slight or No Change'
if((df['Day_Perc_Change'] >= 0.5) & (df['Day_Perc_Change'] <= 1)).all():
    df['Trend'] = 'Slight Positive'
if((df['Day_Perc_Change'] <= -0.5) & (df['Day_Perc_Change'] >= -1)).all():
    df['Trend'] = 'Slight Negative'
if((df['Day_Perc_Change'] >= 1) & (df['Day_Perc_Change'] <= 3)).all():
    df['Trend'] = 'Positive'
if((df['Day_Perc_Change'] >= -1) & (df['Day_Perc_Change'] <= -3)).all():
    df['Trend'] = 'Negative'
if((df['Day_Perc_Change'] >= 3) & (df['Day_Perc_Change'] <= 7)).all():
    df['Trend'] = 'Among Top Gainers'
if((df['Day_Perc_Change'] <= -3) & (df['Day_Perc_Change'] >= -7)).all():
    df['Trend'] = 'Among Top Losers'
if((df['Day_Perc_Change'] > 7)).all():
    df['Trend'] = 'Bull run'
if((df['Day_Perc_Change'] <= -7)).all():
    df['Trend'] = 'Bear Drop'
df[['Day_Perc_Change', 'Trend']]
```
